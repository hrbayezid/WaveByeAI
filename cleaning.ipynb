```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import LabelEncoder
```

```python
df = pd.read_csv("files/Churn.csv")
df.head()
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df["TotalCharges"] = pd.to_numeric(df["TotalCharges"],errors="coerce").astype(float)
print(df["TotalCharges"].isnull().sum())

df["TotalCharges"]  =  df["TotalCharges"].fillna(df["TotalCharges"].mean())
print(df["TotalCharges"].isnull().sum())

print(df["TotalCharges"].dtype)
```

```python
for col in df.columns :
    if df[col].dtype == object:
        print(col , ":" ,df[col].unique())
```

```python
no_service = ["MultipleLines","OnlineSecurity","OnlineBackup","DeviceProtection","TechSupport","StreamingTV","StreamingTV"]

for col in no_service :
    df[col] = df[col].replace({"No phone service" : "No" , "No internet service" : "No"})
```

```python
tnr = df.pop("tenure")
df.insert(17,"tenure",tnr)

target = df.pop("Churn")
df.insert(17,"Churn",target)


df.head()
```

```python
le = LabelEncoder()

catagorical = df.loc[:,"gender":"Churn"]

for col in catagorical:
    df[col] = le.fit_transform(df[col])
df.head()
```

```python
df.to_csv("files/cleaned_Churn.csv",index=False)
```
